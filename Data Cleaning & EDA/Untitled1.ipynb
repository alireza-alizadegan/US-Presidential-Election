```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3])
plt.show
```

```python
import matplotlib.gridspec as gridspec

# gridspec.GridSpec(1,2)
shape = (2,2)

plt.subplot2grid(shape,(0,0),1,2)
plt.plot([1, 2, 3])
plt.show

plt.subplot2grid(shape,(1,0))
plt.plot([1, 2, 3])
plt.show

plt.subplot2grid(shape,(1,1))
plt.plot([1, 2, 3])
plt.show

plt.tight_layout()
```
